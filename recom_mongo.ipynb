## MongoDB installing in local
ref: https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-ubuntu/ ,


Follow [this instruction](https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-ubuntu/) to install mongoDB in local.

To start or stop DB, in terminal
```bash
sudo systemctl start mongod
sudo systemctl stop mongod
```

To check MongoDB status,
```bash
sudo systemctl status mongod
```
if it *active*, it is installed and enabled successfully.

## (Optional) Install Compass (GUI tool)
ref: https://www.mongodb.com/docs/compass/current/install/

*make sure the package is newest version*. You can go https://www.mongodb.com/try/download/compass to copy the newest package link.
- Install

`curl -O -L https://downloads.mongodb.com/compass/mongodb-compass_<ver.>_amd64.deb`

`sudo dpkg -i mongodb-compass_1.43.0_amd64.deb`
- If you have compatibility error, you may use this command below :

`sudo apt --fix-broken install`

`sudo apt-get install libasound2`

- Remove installer file

`rm ./mongodb-compass_<ver.>_amd64.deb`

- To run,

`mongodb-compass`
